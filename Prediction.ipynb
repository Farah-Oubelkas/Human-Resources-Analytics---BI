```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
from sklearn.model_selection import train_test_split 
```

```python
matplotlib notebook
```

```python
df = pd.read_excel (r'C:/Users/Farah/Desktop/Ensa/S8/HumainRessourceApplication/prediction.xlsx')
```

```python
print (df)
```

```python
 %matplotlib inline
df.plot(kind='scatter',x='total_experience',y='compensation')
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.total_experience,df.compensation)
```

```python
plt.scatter(X_train,y_train,label="training data",color='r',alpha=.7)
plt.scatter(X_test,y_test,label="testing data",color='g',alpha=.7)
plt.legend()
plt.title('Test train split')
plt.show()
```

```python
df.plot(kind='scatter',x='total_experience',y='compensation')
```

```python
#create linear model and train it 
LR=LinearRegression()
LR.fit(X_train.values.reshape(-1,1),y_train.values)
```

```python
prediction=LR.predict(X_test.values.reshape(-1,1))
plt.plot(X_test,prediction,label='Linear regression',color='b')
plt.scatter(X_test,y_test, label='Actual test data',color='black',alpha=.7)
plt.legend()
plt.show()
```
